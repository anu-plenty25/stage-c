```python
"""Due to the direct relationship between the 'stab' and 'stabf' column, 'stab' column will be droped and 'stabf'
becomes the sole dependent variable """
```

```python
df.drop('stab', axis = 1,inplace = True)
```

```python
#replacing stable with 1 and unstable with 0 
df['stabf'] = df['stabf'].replace({'unstable': 0,
                   'stable': 1})
```

```python
#Splitting the dataset into train and test 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop('stabf', axis = 1)
y = df['stabf']

scaler = StandardScaler()
scaled_x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(scaled_x, y, test_size = 0.2, random_state = 1)
```

```python
x_test = scaler.transform(x_test)
```

```python
from sklearn.ensemble import ExtraTreesClassifier, RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, confusion_matrix

rf = RandomForestClassifier(random_state = 1)
etc = ExtraTreesClassifier()
```

```python
rf.fit(x_train, y_train)
y_pred = rf.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
etc.fit(x_train, y_train)
y_preds = etc.predict(x_test)
```

```python
params ={'n_estimators':[1000, 300, 500,100],
        'bootstrap':[True, False],
        'min_samples_split':[1,3,5]}
```

```python
from sklearn.model_selection import RandomizedSearchCV
etc_search = RandomizedSearchCV(etc,param_distributions = params, random_state = 1,cv = 5, n_iter = 10, scoring = 'accuracy', n_jobs = -1, verbose = 1)
etc_search.fit(x_train, y_train)
etc_search.best_params_
```

```python
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

xgb = XGBClassifier(random_state = 1)
lgb = LGBMClassifier(random_state = 1)
```

```python
xgb.fit(x_train, y_train)
y_red = xgb.predict(x_test)
accuracy_score(y_test, y_red)
```

```python
lgb.fit(x_train, y_train)
ypred = lgb.predict(x_test)
accuracy_score(y_test, ypred)
```
